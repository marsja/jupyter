## Scatter Plots in Python Using Seaborn
This is the Jupyter Notebook for the Python data visualization tutorial [How to Make a Scatter Plot in Python]( Pythonhttps://www.marsja.se/how-to-make-a-scatter-plot-in-python-using-seaborn/). Here you will learn how to use Seaborns scatterplot method to create scatter plots. 

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

# Import Data
data = 'https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv'
df = pd.read_csv(data, index_col=0)
```

## How to use Seaborn to create Scatter Plots in Python

In the first section of this Seaborn scatter plot tutorial we are going to use the method *scatterplot*.

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='wt', y='mpg', data=df)
```

```python
### Size of the dots

sns.scatterplot(x='wt', y='mpg', 
                size='wt', data=df)
```

```python
### Change the Axis
ax = sns.scatterplot(x='wt', y='mpg', size='wt', data=df)
ax.set(xticks=np.arange(1, 6, 1),
      yticks=np.arange(5, 50, 10))
```

### Grouped Scatter Plot


```python
sns.scatterplot(x='wt', y='mpg',
                hue='vs', data=df)
```

### Markers


```python
data = 'https://vincentarelbundock.github.io/Rdatasets/csv/carData/Burt.csv'
df = pd.read_csv(data, index_col=0)

sns.scatterplot(x='IQbio', y='IQfoster', style='class',
                hue='class', data=df)
```

```python
### Bivariate Distribution on Scatter plot

data = 'https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv'
df = pd.read_csv(data, index_col=0)

sns.scatterplot(x='wt', y='mpg', data=df)
sns.kdeplot(df.wt, df.mpg)
```

### Seaborn Scatter plot using the regplot method
In this section we will use the Seaborn regplot method to create a scatter plot. This enables us to create a scatter plot with a regression line added.

```python
import seaborn as sns

sns.regplot(x='wt', y='mpg', data=df)
```

### Scatter Plot Without Confidence Interval

```python
import seaborn as sns

sns.regplot(x='wt', y='mpg', 
            ci=None, data=df)
```

### Removing the Regression line
If we want to, we can also remove the regression line, from the scatter plot, by using the argument fit_reg and set it to False:

```python
import numpy as np

sns.regplot(x='wt', y='mpg', 
            fit_reg=False, data=df)
```

### Changing the number of Ticks

```python
import seaborn as sns

ax = sns.regplot(x='wt', y='mpg', data=df)
ax.set(xticks=np.arange(2, 6, 1),
      yticks=np.arange(5, 50, 10))
```

## Changing the Color on the Scatterplot

```python
sns.regplot(x='wt', y='mpg', 
            color='g', data=df)
```

### Adding Correlation to the Scatter Plot

```python
from scipy.stats import pearsonr

corr = pearsonr(df['wt'], df['mpg'])
corr = [np.round(c, 2) for c in corr]
print(corr)
```

```python
text = 'r=%s, p=%s' % (corr[0], corr[1])
plt.figure(figsize=(10, 6))

ax = sns.regplot(x="wt", y="mpg", data=df)
ax.text(4, 30, text, fontsize=12)
```

### Seaborn Scatter Plot with Density

```python
import seaborn as sns

# Using regplot to create the scatter plot with 95% confidence interval and a regression line
sns.regplot(x='wt', y='mpg', ci=None, data=df)
sns.kdeplot(df.wt, df.mpg)
```

### How to Rotate the Axis Labels on a Seaborn Plot

```python
ax = sns.regplot(x='wt', y='mpg', data=df)
for item in ax.get_xticklabels():
    item.set_rotation(90)
```
